```python
# Count Negative Numbers in Array

# Without function
n = int(input("Enter number of elements: "))
arr = list(map(int, input("Enter elements: ").split()))
count = 0
for num in arr:
    if num < 0:
        count += 1
print("Negative count:", count)


# With function
def count_negative(arr):
    return sum(1 for num in arr if num < 0)

n = int(input("Enter number of elements: "))
arr = list(map(int, input("Enter elements: ").split()))
print("Negative count:", count_negative(arr))
```

```python
# Remove Vowels from String

# Without function
s = input("Enter a string: ")
vowels = "aeiouAEIOU"
result = ""
for ch in s:
    if ch not in vowels:
        result += ch
print("After removing vowels:", result)


# With function
def remove_vowels(s):
    return "".join(ch for ch in s if ch.lower() not in "aeiou")

s = input("Enter a string: ")
print("After removing vowels:", remove_vowels(s))
```

```python
# Count special characters (not alphabet, digit, space)

# Without function
s = input("Enter string: ")
count = 0
for ch in s:
    if not (ch.isalnum() or ch == " "):
        count += 1
print("Special characters:", count)


# With function
def count_special(s):
    return sum(1 for ch in s if not (ch.isalnum() or ch == " "))

s = input("Enter string: ")
print("Special characters:", count_special(s))
```

```python
# Missing character

# Without function
s1 = input("Enter first string: ")
s2 = input("Enter second string: ")
for ch in s1:
    if ch not in s2 or s1.count(ch) != s2.count(ch):
        print("Missing:", ch)
        break


# With function
def missing_char(s1, s2):
    for ch in s1:
        if ch not in s2 or s1.count(ch) != s2.count(ch):
            return ch
    return None

s1 = input("Enter first string: ")
s2 = input("Enter second string: ")
print("Missing:", missing_char(s1, s2))
```

```python
# Minimum cable length problem

# Without function
n = int(input("Enter no. of systems: "))
state = list(map(int, input("Enter system states: ").split()))
m = int(input("Enter no. of distances: "))
dist = list(map(int, input("Enter distances: ").split()))

total = 0
for i in range(1, n):
    if state[i] == 0:
        total += dist[i] - dist[i-1]
print("Minimum cable length:", total)


# With function
def min_cable(state, dist):
    total = 0
    for i in range(1, len(state)):
        if state[i] == 0:
            total += dist[i] - dist[i-1]
    return total

n = int(input("Enter no. of systems: "))
state = list(map(int, input("Enter system states: ").split()))
m = int(input("Enter no. of distances: "))
dist = list(map(int, input("Enter distances: ").split()))
print("Minimum cable length:", min_cable(state, dist))
```

```python
# Max bits to flip (convert P → Q)

# Without function
p = int(input("Enter P: "))
q = int(input("Enter Q: "))
x = p ^ q
count = bin(x).count("1")
print("Bits to flip:", count)


# With function
def bit_flips(p, q):
    return bin(p ^ q).count("1")

p = int(input("Enter P: "))
q = int(input("Enter Q: "))
print("Bits to flip:", bit_flips(p, q))
```

```python
# Sum of digits of each array element

# Without function
arr = list(map(int, input("Enter array: ").split()))
res = []
for num in arr:
    digit_sum = sum(int(d) for d in str(num))
    res.append(digit_sum)
print("Digit sums:", res)


# With function
def digit_sums(arr):
    return [sum(int(d) for d in str(num)) for num in arr]

arr = list(map(int, input("Enter array: ").split()))
print("Digit sums:", digit_sums(arr))
```

```python
# Count non-repeating characters in string

# Without function
s = input("Enter string: ")
count = 0
for ch in s:
    if s.count(ch) == 1:
        count += 1
print("Non-repeating count:", count)


# With function
def non_repeating_count(s):
    return sum(1 for ch in s if s.count(ch) == 1)

s = input("Enter string: ")
print("Non-repeating count:", non_repeating_count(s))
```

```python
# Sum of numbers with maximum product (positive & negative mix)

# Without function
arr = list(map(int, input("Enter array: ").split()))
arr.sort()
prod1 = arr[-1] * arr[-2]
prod2 = arr[0] * arr[1]
if prod1 > prod2:
    print("Sum:", arr[-1] + arr[-2])
else:
    print("Sum:", arr[0] + arr[1])


# With function
def max_product_sum(arr):
    arr.sort()
    prod1 = arr[-1] * arr[-2]
    prod2 = arr[0] * arr[1]
    return arr[-1] + arr[-2] if prod1 > prod2 else arr[0] + arr[1]

arr = list(map(int, input("Enter array: ").split()))
print("Sum:", max_product_sum(arr))
```

```python
#First negative number in each sliding window of size k

#  Without function
arr = list(map(int, input("Enter array: ").split()))
k = int(input("Enter window size: "))
n = len(arr)
res = []
for i in range(n-k+1):
    window = arr[i:i+k]
    neg = 0
    for num in window:
        if num < 0:
            neg = num
            break
    res.append(neg)
print("First negatives:", res)


# With function
def first_negatives(arr, k):
    res = []
    n = len(arr)
    for i in range(n-k+1):
        neg = 0
        for num in arr[i:i+k]:
            if num < 0:
                neg = num
                break
        res.append(neg)
    return res

arr = list(map(int, input("Enter array: ").split()))
k = int(input("Enter window size: "))
print("First negatives:", first_negatives(arr, k))
```

```python
# A given Series could be arithmetic,Geometric,Fibonacci.We will provide you 'n' numbers and you have to first check the type of series,then find out the next number .If you are unable to get the output return -999.
# Input1 : integer value,the length of the array. Input2 : Integer Array e.g(0,1,2,3)

def next_in_series(arr):
    n = len(arr)

    # Check for Arithmetic Progression (AP)
    diff = arr[1] - arr[0]
    if all(arr[i] - arr[i-1] == diff for i in range(1, n)):
        return arr[-1] + diff

    # Check for Geometric Progression (GP)
    if arr[0] != 0:  # avoid division by zero
        ratio = arr[1] // arr[0]
        if all(arr[i] * arr[0] == arr[1] * arr[i-1] for i in range(1, n)):
            return arr[-1] * ratio

    # Check for Fibonacci
    if n >= 3 and all(arr[i] == arr[i-1] + arr[i-2] for i in range(2, n)):
        return arr[-1] + arr[-2]

    return -999


# Input
n = int(input("Enter length of series: "))
arr = list(map(int, input("Enter series: ").split()))

print("Next number:", next_in_series(arr))
```
